```python
train.columns
```

```python
trainML = train[['Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Embarked', 'Name_len'
                , 'Ticket_First', 'FamilyCount', 'Title']]
```

### Logistic Regression - Age

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
X_Age = trainML[['Age']].values
```

```python
y = trainML['Survived'].values
```

```python
classifier.fit(X_Age, y)
```

```python
y_pred = classifier.predict(X_Age)
```

```python
y_pred[:10]
```

```python
(y == y_pred).mean()
```

### Logistic Regression - Fare

```python
X_Fare = trainML[['Fare']].values
```

```python
classifier.fit(X_Fare,y)
```

```python
y_pred = classifier.predict(X_Fare)
```

```python
(y == y_pred).mean()
```
